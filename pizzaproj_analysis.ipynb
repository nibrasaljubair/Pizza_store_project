```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv('pizzaorder_analysis.csv')
```

```python
# find total sales 
Orde
total_sales = (df['item_price'] * df['quantity']).sum()
```

```python
total_sales
```

```python
# sales by categ0ry(donut chart)

sales_by_category = (df['item_price'] * df['quantity']).groupby(df['item_cat']).sum() / total_sales * 100
```

```python
colors = ['#2D6154', '#35726C', '#419179', '#66B29D', '#92D2BF']

plt.figure(figsize=(8, 8))
wedges,texts,autotexts = plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%', startangle=140, wedgeprops={'width':0.3}, colors=colors)
plt.title('Sales by Category (% of Total Sales)')
plt.legend(wedges, sales_by_category.index, title="Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

```python
# top selling items 
top_selling_items = df.groupby('item_name').apply(lambda x: (x['item_price'] * x['quantity']).sum()).nlargest(10)


color2 = ['#2D6154', '#35726C', '#419179', '#66B29D', '#92D2BF', '#B7E5D0', '#CFF0E1', '#E5FAF2', '#A9D7C2', '#74B99C']


plt.figure(figsize=(12, 6))
top_selling_items.plot(kind='bar', color=color2[:len(top_selling_items)], edgecolor='black')
plt.xlabel('Item Name')
plt.ylabel('Total Sales ($)')
plt.title('Top 10 Selling Items by Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Orders by Hour and Sales by Hour 
df['created_at'] = pd.to_datetime(df['created_at'])
df['hour'] = df['created_at'].dt.hour


orders_by_hour = df.groupby('hour').size()


sales_by_hour = df.groupby('hour').apply(lambda x: (x['item_price'] * x['quantity']).sum())

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(sales_by_hour.index, sales_by_hour, color=colors[0], linewidth=2, label='Total Sales')
ax1.set_xlabel('Hour of the Day')
ax1.set_ylabel('Total Sales ($)', color=colors[0])
ax1.tick_params(axis='y', labelcolor=colors[0])


ax2 = ax1.twinx()
ax2.plot(orders_by_hour.index, orders_by_hour, color=colors[2], linewidth=2, label='Total Orders')
ax2.set_ylabel('Total Orders', color=colors[2])
ax2.tick_params(axis='y', labelcolor=colors[2])

fig.legend(loc="upper left", bbox_to_anchor=(0.3, 0.9), bbox_transform=ax1.transAxes)

fig.suptitle('Total Sales and Orders by Hour')
ax1.grid(True)

plt.show()
```
